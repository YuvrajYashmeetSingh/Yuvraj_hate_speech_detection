```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
data = pd.read_csv("labeled_data.csv")

# Display the first few rows of the dataset to understand its structure
print(data.head())
```

```python
# Visualize the distribution of 'count' column
sns.histplot(data['count'], kde=True)
plt.title('Distribution of Count')
plt.xlabel('Count')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualize the distribution of 'class' column
sns.countplot(x='class', data=data)
plt.title('Distribution of Classes')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

```python
# Visualize the correlation between 'count', 'hate_speech', 'offensive_language', and 'neither'
correlation_matrix = data[['count', 'hate_speech', 'offensive_language', 'neither']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()

```

```python

# Visualize the relationship between 'count' and 'hate_speech' using a scatter plot
sns.scatterplot(x='count', y='hate_speech', data=data)
plt.title('Relationship between Count and Hate Speech')
plt.xlabel('Count')
plt.ylabel('Hate Speech')
plt.show()
```

```python
# Visualize the distribution of 'tweet' lengths
data['tweet_length'] = data['tweet'].apply(len)
sns.histplot(data['tweet_length'], kde=True)
plt.title('Distribution of Tweet Lengths')
plt.xlabel('Tweet Length')
plt.ylabel('Frequency')
plt.show()
```
